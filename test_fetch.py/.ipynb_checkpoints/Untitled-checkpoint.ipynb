```python
import pandas as pd

# Create a simple DataFrame
df = pd.DataFrame({
    "trade_date": ["2026-06-01", "2026-06-02", "2026-06-03"],
    "close": [2458.20, 2470.30, 2480.50]
})

print(df)
```

```python
# Calculate daily return (percentage change from previous day)
df["daily_return"] = df["close"].pct_change()
print(df)
```

```python
# Calculate 7-day moving average (min_periods=1 means it calculates even with fewer rows)
df["ma_7"] = df["close"].rolling(7, min_periods=1).mean()
print(df)
```

```python
# Save as JSON (one object per line, used for S3 upload)
df.to_json("data.json", orient="records", lines=True)
print("✅ Saved to data.json!")
```

```python
import numpy as np

# Create a numpy array of stock prices
prices = np.array([2458.20, 2470.30, 2480.50, 2478.00, 2488.00])

print("Prices:", prices)
print("Mean:", np.mean(prices))
print("Max:", np.max(prices))
print("Min:", np.min(prices))
print("Std Dev:", np.std(prices))
```

```python
# Daily returns: (today - yesterday) / yesterday
daily_returns = np.diff(prices) / prices[:-1]

print("Daily returns:", daily_returns)
print("Cumulative return:", np.prod(1 + daily_returns) - 1)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np

# 1. Fetch real stock data from Yahoo Finance
ticker = yf.Ticker("RELIANCE.NS")
data = ticker.history(period="5d", interval="1d")

# 2. Transform with pandas
df = data.reset_index()
df["ticker"] = "RELIANCE.NS"
df["daily_return"] = df["Close"].pct_change()
df["volatility"] = (df["High"] - df["Low"]) / df["Open"]
df["ma_7"] = df["Close"].rolling(7, min_periods=1).mean()

# 3. Calculate with numpy
avg_return = np.mean(df["daily_return"].dropna())
print(f"Average daily return: {avg_return:.4f}")

# Show the result
df.round(4)
```
